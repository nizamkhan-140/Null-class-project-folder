```python
import tkinter as tk
from tkinter import messagebox

# Simple dictionary dataset (for demo) – replace with your trained model outputs
en_to_fr = {
    "hello my friend": "bonjour mon ami",
    "i love programming": "j'aime programmer",
    "good morning everyone": "bonjour tout le monde"
}

en_to_hi = {
    "hello my friend": "नमस्ते मेरे दोस्त",
    "i love programming": "मुझे प्रोग्रामिंग पसंद है",
    "good morning everyone": "सुप्रभात सभी को"
}

def translate_text():
    text = entry.get().lower().strip()
    # Check letter count
    letters_only = "".join([c for c in text if c.isalpha()])
    if len(letters_only) < 10:
        messagebox.showwarning("Warning", "upload again")
        return
    
    # Lookup translations
    fr = en_to_fr.get(text, "No translation found")
    hi = en_to_hi.get(text, "No translation found")
    
    output_fr.config(text=f"French: {fr}")
    output_hi.config(text=f"Hindi: {hi}")

# GUI
root = tk.Tk()
root.title("Dual Language Translator")

tk.Label(root, text="Enter English sentence:", font=("Arial", 14)).pack(pady=5)
entry = tk.Entry(root, width=50, font=("Arial", 14))
entry.pack(pady=5)

btn = tk.Button(root, text="Translate", command=translate_text, font=("Arial", 14))
btn.pack(pady=10)

output_fr = tk.Label(root, text="French: ", font=("Arial", 14), fg="blue")
output_fr.pack(pady=5)

output_hi = tk.Label(root, text="Hindi: ", font=("Arial", 14), fg="green")
output_hi.pack(pady=5)

root.mainloop()
```
