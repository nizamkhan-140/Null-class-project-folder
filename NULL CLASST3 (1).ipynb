```python
import tkinter as tk
from tkinter import messagebox
import speech_recognition as sr
from deep_translator import GoogleTranslator
from datetime import datetime

```

```python
#THIS FUNCTION CHECT THE ACTIVE TIME FROM (9:30 PM TO 10:00PM)
def is_active_time():
    current_time = datetime.now().time()
    start_time = datetime.strptime("12:13", "%H:%M").time()
    end_time = datetime.strptime("22:58","%H:%M").time()
    return start_time <= current_time <= end_time
```

```python
#THIS IS THE FUNCTION TO CAPTURE AUDIO FROM MICROPHONE ,TRANSLATION AND GUII DISPLAY
def translate_audio():
    if not is_active_time():
        messagebox.showinfo("Info","Taking rest,see you tomorrow!")
        return
    recognizer = sr.Recognizer()
    with sr.Microphone() as source:
        status_label.config(text="Listening... Speakk in English")
        root.update()
        try:
            audio = recognizer.listen(source, timeout=10, phrase_time_limit=10)
            text = recognizer.recognize_google(audio, language="en-US")
            status_label.config(text=f"Recognized: {text}")
            Translated_text = GoogleTranslator(source='en', target='hi').translate(text)
            
            output_text.delete("1.0", tk.END)
            output_text.insert(tk.END, Translated_text)
        except sr.UnknownValueError:
            messagebox.showerror("Error", "Could not understand audio, please repeat!")
        except sr.RequestError:
            messagebox.showerror("Error", "Network issue. Please check your internet!")
```

```python
#IMPORT TKINTEER TO BULD GUI INTERFACE
root = tk.Tk()
root.title("Voice Translator (English â†’ Hindi)")
root.geometry("500x400")

tk.Label(root, text="Voice Translator", font=("Arial", 16, "bold")).pack(pady=10)
status_label = tk.Label(root, text="Click 'Start Listening' to speak", font=("Arial", 12))
status_label.pack(pady=5)

start_button = tk.Button(root, text="Start Listening", command=translate_audio, font=("Arial", 12), bg="lightgreen")
start_button.pack(pady=20)

tk.Label(root, text="Translated Hindi Text:", font=("Arial", 14)).pack()
output_text = tk.Text(root, height=5, width=50, font=("Arial", 12))
output_text.pack(pady=10)

root.mainloop()
```
